# Formula Used


W= (Rv+Cm)/(v+m)

W=Weighted Rating
R=average for the movie asanumber from0to 10(mean)=(Rating)
V=number of votes for the movie-(votes)
M=minimum votes required to be listed in the Top 250(currently 3000)
C=the mean vote across the whole report(currently 6.9)

where:
Source:http://trailerpark.weebly.com/imdb-rating.html?source=post page
